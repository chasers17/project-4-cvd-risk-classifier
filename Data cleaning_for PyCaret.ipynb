```python
df.age=(df["age"]/365.25).round(0) 
df.head(5)
```

```python
# Calculate IQR for 'ap_hi' and 'ap_lo'
Q1_hi = df['ap_hi'].quantile(0.25)
Q3_hi = df['ap_hi'].quantile(0.75)
IQR_hi = Q3_hi - Q1_hi
lower_bound_hi = Q1_hi - 1.5 * IQR_hi
upper_bound_hi = Q3_hi + 1.5 * IQR_hi
# Calculate IQR for 'ap_lo'
Q1_lo = df['ap_lo'].quantile(0.25)
Q3_lo = df['ap_lo'].quantile(0.75)
IQR_lo = Q3_lo - Q1_lo
lower_bound_lo = Q1_lo - 1.5 * IQR_lo
upper_bound_lo = Q3_lo + 1.5 * IQR_lo
# Filter out outliers based on 'ap_hi' and 'ap_lo'
filtered_df = df[(df['ap_hi'].between(lower_bound_hi, upper_bound_hi)) &
                        (df['ap_lo'].between(lower_bound_lo, upper_bound_lo))]
```

```python
# df1=df.loc[(df["ap_hi"]<210) | (df["height"]>=140) | (df["weight"]>=40)]
filtered_cardio1=filtered_df.loc[(filtered_df["height"]>=140) & (filtered_df["height"] <=200)]
filtered_cardio=filtered_cardio1.loc[filtered_cardio1["weight"]>=40]
filtered_cardio
```
